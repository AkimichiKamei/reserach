```python
# order_v1

count = 0   
order = []

for i in range(16):
    for j in range(count, 16):
        si = sum(imgList[i])
        sj = sum(imgList[j])
        if si == sj:
            if si == 0 or si == 4:
                order.append(24)
            elif si == 1 or si == 3:
                order.append(6)
            elif si == 2:
                order.append(4)
        else:
            order.append(0)
            
    count += 1
        
    if count == 16:
        break        
```

```python
# order_v2

count = 0
order = []

for i in range(16):
    for j in range(count, 16):
        si = sum(imgList[i])
        sj = sum(imgList[j])
        
        if si == sj:
            if si == 0 or si == 4:
                order.append(24)
            elif si == 1:
                if np.allclose(imgList[i],imgList[j]):
                    order.append(6)
                else:
                    order.append(4)
            elif si == 2:
                if np.allclose(imgList[i]+imgList[j],np.array([1,1,1,1])):
                    order.append(8)
                elif np.allclose(imgList[i],imgList[j]):
                    order.append(4)
                else:
                    order.append(2)
            elif si == 3:
                if np.allclose(imgList[i],imgList[j]):
                    order.append(6)
                else:
                    order.append(4)
        else:
            order.append(0)
            
    count += 1        
    if count == 16:
        break    
```
